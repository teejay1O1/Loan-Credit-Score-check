#### remove data olumns with reaccuring data for all columns

```python
data.drop(columns=["pymnt_plan","initial_list_status","application_type"],inplace=True)
data.shape
```

```python
dummy_columns=pd.get_dummies(data["loan_status"])
data_final=pd.concat([dummy_columns,data],axis=1)
data_final.drop(columns=["loan_status"],inplace=True)
#data_final["loan_status"]=data_final["Charged Off"]*1+data_final["Current"]*2+data_final["Fully Paid"]*3
#data_final.drop(columns=["Charged Off","Current","Fully Paid"],inplce=True)
data.shape
```

```python
data_final.head()
```

```python
correlation=data_final.corr()
correlation.head(3)
```

```python
for i in correlation.columns.values :
    pearson_coef_CF, p_value_CF = stats.pearsonr(data_final[i], data_final["Charged Off"])
    pearson_coef_Cu, p_value_Cu = stats.pearsonr(data_final[i], data_final["Current"])
    pearson_coef_CF, p_value_FP = stats.pearsonr(data_final[i], data_final["Fully Paid"])
    
    if(p_value_CF>0.05 and p_value_Cu>0.05 and p_value_FP>0.1):
        print("The Pearson Correlation Coefficient of",i)
```

#### droping data which have negligible correlation with loan status
