```python
excel= pd.ExcelFile(file1)
```

```python
data=excel.parse(excel.sheet_names[0])
```

### Data Wrangling

```python
newdata=data.dropna(axis=1,how="all")
```

```python
newdata.head()
```

### Grouping

```python
grouped=newdata.groupby(["loan_status"])
```

```python
newgrouped=grouped.count()
newgrouped.head()
```

### Data Visuals

```python
plt.bar([1,2,3],newgrouped['loan_amnt'])
plt.ylabel("Amount")
plt.xlabel("Status Of Loan")
plt.title("Dependence on amount of loan")
plt.xticks([1,2,3],["Charged","current","FP"])
plt.show()
```
